```python
# ========================================
# Step 4: PDF Text Extraction & Corpus Construction
# ========================================

import os
import re
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import PyPDF2
import pdfplumber
from langdetect import detect, LangDetectException
import warnings
warnings.filterwarnings('ignore')

# Paths
BASE_PATH = '/content/drive/My Drive/MasterThesis_Archetypes'
PDF_PATH = os.path.join(BASE_PATH, '00_RawData/Macro_PDFs')
OUTPUT_PATH = os.path.join(BASE_PATH, '01_ProcessedCorpora')

print("="*70)
print("📖 PDF TEXT EXTRACTION ENGINE")
print("="*70)

# ========================================
# IDEOLOGY MAPPING RULES
# ========================================
# Based on your thesis MECE categories

IDEOLOGY_MAP = {
    # Extreme Left
    'Karl-Marx': 'Extreme_Left',
    'Friedrich-Engels': 'Extreme_Left',
    'Lenin': 'Extreme_Left',
    'V-I-Lenin': 'Extreme_Left',
    'Stalin': 'Extreme_Left',
    'J-V-Stalin': 'Extreme_Left',
    'Mao-Tse-tung': 'Extreme_Left',
    'Fidel-Castro': 'Extreme_Left',
    'Hugo-Chavez': 'Extreme_Left',
    'Gustavo-Petro': 'Extreme_Left',
    'RAF': 'Extreme_Left',
    'Mikhail-Bakunin': 'Extreme_Left',

    # Extreme Right
    'Adolf-Hitler': 'Extreme_Right',
    'Benito-Mussolini': 'Extreme_Right',
    'Giovanni-Gentile': 'Extreme_Right',
    'Ernst-Hiemer': 'Extreme_Right',
    'AfD': 'Extreme_Right',
    'Meir-Kahane': 'Extreme_Right',
    'Yitzhak-Shapira': 'Extreme_Right',
    'Christchurch-Shooter': 'Extreme_Right',
    'Eric-Harris': 'Extreme_Right',
    'Dylan-Klebold': 'Extreme_Right',

    # Jihadist
    'Hamas': 'Jihadist',
    'Osama-bin-Laden': 'Jihadist',

    # Mainstream Democratic
    'CDU': 'Mainstream_Democratic',
    'SPD': 'Mainstream_Democratic',
    'European-Commission': 'Mainstream_Democratic',
    'Global-Policy-Forum': 'Mainstream_Democratic',
    'WOLA': 'Mainstream_Democratic',

    # Libertarian (5th category)
    'Friedrich-Hayek': 'Libertarian',
    'Milton-Friedman': 'Libertarian',
    'Murray-Rothbard': 'Libertarian',
    'La-Libertad-Avanza': 'Libertarian',

    # Ambiguous (will assign manually)
    'Die-Linke': 'Extreme_Left',  # German far-left party
}

def assign_ideology(author):
    """Assign ideology based on author name"""
    for key, ideology in IDEOLOGY_MAP.items():
        if key in author:
            return ideology
    return 'MANUAL_REVIEW'  # Flag for manual classification

# ========================================
# EXTRACTION FUNCTIONS
# ========================================

def extract_with_pypdf2(pdf_path):
    """Primary extraction method"""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page in reader.pages:
                text += page.extract_text() + "\n"
            return text.strip()
    except Exception as e:
        return None

def extract_with_pdfplumber(pdf_path):
    """Fallback extraction method"""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text() + "\n"
            return text.strip()
    except Exception as e:
        return None

def extract_text(pdf_path):
    """Smart extraction with fallback"""
    text = extract_with_pypdf2(pdf_path)
    if not text or len(text) < 100:  # If extraction failed or too short
        text = extract_with_pdfplumber(pdf_path)
    return text if text else ""

def clean_text(text):
    """Basic text cleaning"""
    # Remove excessive whitespace
    text = re.sub(r'\s+', ' ', text)
    # Remove page numbers (basic pattern)
    text = re.sub(r'\n\d+\n', '\n', text)
    # Remove form feed characters
    text = text.replace('\f', '\n')
    return text.strip()

def detect_language_safe(text):
    """Detect language with fallback"""
    try:
        # Use first 1000 chars for detection
        sample = text[:1000] if len(text) > 1000 else text
        return detect(sample)
    except LangDetectException:
        return 'unknown'

def parse_filename(filename):
    """Parse filename with double underscores"""
    # Remove .pdf extension
    name = filename.replace('.pdf', '')

    # Split by double underscores
    parts = name.split('__')

    metadata = {
        'year': parts[0] if len(parts) > 0 else 'UNKNOWN',
        'author': parts[1] if len(parts) > 1 else 'UNKNOWN',
        'doc_type': parts[2] if len(parts) > 2 else 'UNKNOWN',
        'title': parts[3] if len(parts) > 3 else 'UNKNOWN',
        'language_code': parts[4] if len(parts) > 4 else 'UNKNOWN'
    }

    return metadata

# ========================================
# MAIN EXTRACTION LOOP
# ========================================

print("\n🔄 Starting extraction...\n")

pdf_files = sorted([f for f in os.listdir(PDF_PATH) if f.endswith('.pdf')])
corpus_data = []

for filename in tqdm(pdf_files, desc="📄 Extracting PDFs", ncols=70):
    pdf_path = os.path.join(PDF_PATH, filename)

    # Parse metadata
    metadata = parse_filename(filename)

    # Extract text
    text = extract_text(pdf_path)

    # Clean text
    text_clean = clean_text(text)

    # Quality metrics
    word_count = len(text_clean.split())
    char_count = len(text_clean)
    extraction_quality = 'GOOD' if word_count > 100 else 'POOR'

    # Language detection
    detected_lang = detect_language_safe(text_clean)

    # Ideology assignment
    ideology = assign_ideology(metadata['author'])

    # Store data
    corpus_data.append({
        'doc_id': len(corpus_data) + 1,
        'filename': filename,
        'year': metadata['year'],
        'author': metadata['author'],
        'doc_type': metadata['doc_type'],
        'title': metadata['title'],
        'language_code': metadata['language_code'],
        'detected_language': detected_lang,
        'ideology': ideology,
        'word_count': word_count,
        'char_count': char_count,
        'extraction_quality': extraction_quality,
        'text': text_clean
    })

# ========================================
# CREATE DATAFRAME & SAVE
# ========================================

print("\n📊 Creating corpus dataframe...\n")
df = pd.DataFrame(corpus_data)

# Save to CSV
output_file = os.path.join(OUTPUT_PATH, 'macro_corpus_clean.csv')
df.to_csv(output_file, index=False, encoding='utf-8')

print("="*70)
print("✅ EXTRACTION COMPLETE!")
print("="*70)

# ========================================
# SUMMARY STATISTICS
# ========================================

print("\n📈 CORPUS STATISTICS:\n")
print(f"Total documents: {len(df)}")
print(f"Total words: {df['word_count'].sum():,}")
print(f"Average words per document: {df['word_count'].mean():.0f}")
print(f"\n📊 By Ideology:")
print(df['ideology'].value_counts())
print(f"\n🌍 By Language (detected):")
print(df['detected_language'].value_counts())
print(f"\n✅ Extraction Quality:")
print(df['extraction_quality'].value_counts())

# Show documents needing manual review
manual_review = df[df['ideology'] == 'MANUAL_REVIEW']
if len(manual_review) > 0:
    print(f"\n⚠️  {len(manual_review)} documents need manual ideology assignment:")
    for _, row in manual_review.iterrows():
        print(f"   - {row['author']} ({row['year']})")

# Show poor extractions
poor_quality = df[df['extraction_quality'] == 'POOR']
if len(poor_quality) > 0:
    print(f"\n⚠️  {len(poor_quality)} documents had poor extraction (< 100 words):")
    for _, row in poor_quality.iterrows():
        print(f"   - {row['filename']}")

print(f"\n💾 Saved to: {output_file}")
print("="*70)

# Display first 3 rows
print("\n📋 PREVIEW (First 3 documents):\n")
print(df[['doc_id', 'year', 'author', 'ideology', 'detected_language', 'word_count']].head(3))
```

```python
# ========================================
# NB1.5: Deep Text Cleaning (CONSERVATIVE)
# ========================================

import pandas as pd
import re
import os

# Load ORIGINAL corpus (not the broken cleaned one)
BASE_PATH = '/content/drive/My Drive/MasterThesis_Archetypes'
df = pd.read_csv(os.path.join(BASE_PATH, '01_ProcessedCorpora/macro_corpus_clean.csv'), encoding='utf-8')

print("="*70)
print("🧹 CONSERVATIVE TEXT CLEANING")
print("="*70)

def conservative_clean_text(text):
    """
    Targeted cleaning - only fix obvious artifacts
    """
    if not isinstance(text, str):
        return ""

    # 1. Fix ONLY single-letter spacing artifacts (capital letter case)
    # Pattern: "K ommunistischen" → "Kommunistischen"
    # Match: Capital letter + space + lowercase letter
    text = re.sub(r'\b([A-ZÀÂÄÆÇÈÉÊËÌÍÎÏÑÒÓÔÖÙÚÛÜ])\s+([a-zàâäæçèéêëìíîïñòóôöùúûü])', r'\1\2', text)

    # 2. Fix single isolated letters before words (OCR artifact)
    # Pattern: "I n h a l t" → "I nhalt" (then next pass catches "In halt" etc.)
    # But be careful not to remove legitimate single letters like "a" "I"
    text = re.sub(r'\b([a-zàâäæçèéêëìíîïñòóôöùúûü])\s+(?=[a-zàâäæçèéêëìíîïñòóôöùúûü]{3,})', r'\1', text)

    # 3. Remove timestamps (PDF metadata)
    text = re.sub(r'\d{1,2}/\d{1,2}/\d{2,4},?\s*\d{1,2}:\d{2}\s*[AP]M', '', text)

    # 4. Remove document page markers
    text = re.sub(r'\(Document page \d+\)', '', text)
    text = re.sub(r'Document \d+ of \d+', '', text)

    # 5. Remove page numbers on separate lines
    text = re.sub(r'^\d+$', '', text, flags=re.MULTILINE)
    text = re.sub(r'\n\d+\n', '\n', text)

    # 6. Remove bullet points and form feeds
    text = text.replace('•', '')
    text = text.replace('', '')
    text = text.replace('\f', '\n')

    # 7. Normalize dashes and quotes
    text = re.sub(r'[–—―]', '-', text)
    text = re.sub(r'[''`´]', "'", text)
    text = re.sub(r'[""«»]', '"', text)

    # 8. Remove excessive whitespace (but preserve single spaces!)
    text = re.sub(r' {2,}', ' ', text)  # Only multiple spaces
    text = re.sub(r'\n{3,}', '\n\n', text)

    # 9. Strip leading/trailing whitespace per line
    lines = text.split('\n')
    lines = [line.strip() for line in lines]
    text = '\n'.join(lines)

    return text.strip()

# Apply cleaning
print("\n🔄 Applying conservative cleaning to all documents...\n")
from tqdm import tqdm

df['text_clean'] = [conservative_clean_text(text) for text in tqdm(df['text'], desc="Cleaning", ncols=70)]

# Recalculate word count
df['word_count_clean'] = df['text_clean'].apply(lambda x: len(x.split()))

# Compare before/after
print("\n" + "="*70)
print("📊 CLEANING IMPACT")
print("="*70)

total_words_before = df['word_count'].sum()
total_words_after = df['word_count_clean'].sum()
words_removed = total_words_before - total_words_after

print(f"\nTotal words BEFORE: {total_words_before:,}")
print(f"Total words AFTER:  {total_words_after:,}")
print(f"Words removed:      {words_removed:,} ({words_removed/total_words_before*100:.1f}%)")

# Quality check: flag documents with suspicious drops
print("\n⚠️  Quality Check: Documents with >20% word loss:")
df['word_loss_pct'] = (df['word_count'] - df['word_count_clean']) / df['word_count'] * 100
suspicious = df[df['word_loss_pct'] > 20].sort_values('word_loss_pct', ascending=False)

if len(suspicious) > 0:
    print(f"\n{len(suspicious)} documents flagged:")
    for _, row in suspicious.head(5).iterrows():
        print(f"  - {row['filename'][:50]}: {row['word_count']:,} → {row['word_count_clean']:,} ({row['word_loss_pct']:.1f}% loss)")
else:
    print("\n✅ No suspicious drops detected!")

# Show sample improvement
print("\n" + "="*70)
print("🔍 BEFORE/AFTER COMPARISON")
print("="*70)

sample_idx = 0
sample = df.iloc[sample_idx]

print(f"\nDocument: {sample['filename'][:60]}")
print(f"Words: {sample['word_count']:,} → {sample['word_count_clean']:,}")
print("\n--- BEFORE (first 500 chars) ---")
print(sample['text'][:500])
print("\n--- AFTER (first 500 chars) ---")
print(sample['text_clean'][:500])

# Save cleaned version
output_file = os.path.join(BASE_PATH, '01_ProcessedCorpora/macro_corpus_CLEANED.csv')
df_clean = df[['doc_id', 'filename', 'year', 'author', 'doc_type', 'title',
               'language_code', 'detected_language', 'ideology',
               'word_count_clean', 'char_count', 'extraction_quality', 'text_clean']]
df_clean.columns = ['doc_id', 'filename', 'year', 'author', 'doc_type', 'title',
                    'language_code', 'detected_language', 'ideology',
                    'word_count', 'char_count', 'extraction_quality', 'text']

df_clean.to_csv(output_file, index=False, encoding='utf-8')

print("\n" + "="*70)
print("✅ CONSERVATIVE CLEANING COMPLETE!")
print("="*70)
print(f"\n💾 Cleaned corpus saved to:")
print(f"   {output_file}")
```
